```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Step 1: Data Loading
train_data = pd.read_csv('C:/Users/ASUS 502/Downloads/p1_test.csv')
test_data = pd.read_csv('C:/Users/ASUS 502/Downloads/p1_train.csv')
```

```python
train_data.head(5)
```

```python
train_data.columns=["Feature1","Feature2","Target_value"]
```

```python
train_data.head()
```

```python
# Separate features and target variable
X_train = train_data.iloc[:, :-1]
y_train = train_data.iloc[:, -1]
```

```python
test_data.head()
```

```python
test_data.columns=["Feature1","Feature2","Target_value"]
```

```python
test_data.head()
```

```python
# Separate features and target variable
X_test = test_data.iloc[:, :-1]
y_test = test_data.iloc[:, -1]
```

```python
# Reshape the feature data for training and test sets
X_train_reshaped = X_train.values.reshape(-1, 1)
X_test_reshaped = X_test.values.reshape(-1, 1)
```

```python
X_train_reshaped = X_train_reshaped[:len(y_train)]
```

```python
# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train_reshaped, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test_reshaped)
```

```python
# Trim y_pred to match the number of samples in y_test
y_pred = y_pred[:len(y_test)]
```

```python
# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```

```python
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```
